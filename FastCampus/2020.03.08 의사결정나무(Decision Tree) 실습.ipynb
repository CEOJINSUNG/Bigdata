# 2. 의사결정나무 구축 및 시각화

- 트리 구축

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(iris.data, iris.target)
```

- 트리의 시각화

```python
dot_data = tree.export_graphviz(clf, out_file = None, feature_names = iris.feature_names, class_names = iris.target_names, 
                                filled = True, rounded = True, special_characters = True)
graph = graphviz.Source(dot_data)
```

```python
graph
```

- 엔트로피를 활용한 트리

```python
clf2 = tree.DecisionTreeClassifier(criterion = "entropy")
```

```python
clf2.fit(iris.data, iris.target)
```

```python
dot_data2 = tree.export_graphviz(clf2, out_file = None, feature_names = iris.feature_names, class_names = iris.target_names, 
                                filled = True, rounded = True, special_characters = True)
graph2 = graphviz.Source(dot_data2)
```

```python
graph2
```

- 프루닝(과적합을 방지하기 위해서 프루닝을 진행한다. max_depth로 조절함)

```python
clf3 = tree.DecisionTreeClassifier(criterion = "entropy", max_depth = 2)
```

```python
clf3.fit(iris.data, iris.target)
```

```python
dot_data3 = tree.export_graphviz(clf3, out_file = None, feature_names = iris.feature_names, class_names = iris.target_names, 
                                filled = True, rounded = True, special_characters = True)
graph3 = graphviz.Source(dot_data3)
graph3
```

- Confusion Matrix 구하기

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(iris.target,clf.predict(iris.data))
```

# 3. Training - Test 구분 및 Confusion matrix 계산

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, stratify = iris.target, random_state = 1)
```

```python
clf4 = tree.DecisionTreeClassifier(criterion = "entropy")
```

```python
clf4.fit(X_train, y_train)
```

```python
confusion_matrix(y_test, clf4.predict(X_test))
```

# 4. Decision regression tree

- 모듈 불러오기 및 데이터 생성

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt

rng = np.random.RandomState(1)
X = np.sort(5 * rng.rand(80, 1), axis=0)
y = np.sin(X).ravel()
y[::5] += 3 * (0.5 - rng.rand(16))
```

- Regression tree 구축

```python
regr1 = tree.DecisionTreeRegressor(max_depth = 2)
regr2 = tree.DecisionTreeRegressor(max_depth = 5)
```

```python
regr1.fit(X, y)
```

```python
regr2.fit(X, y)
```

```python
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
X_test
```

```python
y_1 = regr1.predict(X_test)
y_2 = regr2.predict(X_test)
```

```python
y_1
```

```python
plt.figure()
plt.scatter(X, y, s=20, edgecolor="black",
            c="darkorange", label="data")
plt.plot(X_test, y_1, color="cornflowerblue",
         label="max_depth=2", linewidth=2)
plt.plot(X_test, y_2, color="yellowgreen", label="max_depth=5", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```

```python
dot_data4 = tree.export_graphviz(regr2, out_file=None, 
                                filled=True, rounded=True,  
                                special_characters=True)
```

```python
graph4 = graphviz.Source(dot_data4) 
graph4
```
