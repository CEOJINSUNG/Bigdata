###### request 모듈
1. http request/response를 위한 모듈
2. HTTP method를 메소드 명으로 사용하여 request 요청 예) get, post

```python
import requests
```

###### get 요청하기

1. http post 요청하기
2. post data 이용하여 데이터 전달하기

```python
url='https://news.naver.com/main/ranking/read.nhn?mid=etc&sid1=111&rankingType=popular_day&oid=092&aid=0002178438&date=20200110&type=1&rankingSeq=10&rankingSectionId=105'
resp=requests.get(url)

resp.text
```

###### post 요청하기
1. http post 요청하기
2. post data 이용하여 데이터 전달하기

```python
url = 'https://xo.nate.com/Notfound.sk?redirect=https%3A%2F%2Fwww.nate.com%2F&cpurl=www_ndr.nate.com%252Flogin%252F&viewnid=&flag=Y&e_url=http%3A%2F%2Fwww.nate.com%2F&mode='
data = {
    'id':'test',
    'pwd':'1234'
}
resp = requests.post(url, data=data)
resp.text
```
