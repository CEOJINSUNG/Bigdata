```python
# matplotlib.pyplot에서 pyplot는 matplotlib의 함수를 불러오는 하나의 API라고 볼수 있음
# 여기서 [1,2,3,4]는 ylabel이 선언되었으므로 여기서 y축의 요소라고 볼수 있다
import matplotlib.pyplot as plt
plt.plot([1,2,3,4])
plt.ylabel('some numbers')
plt.show()
```

```python
import numpy as np

# 0에서 5까지 0.2 간격으로 생성
t = np.arange(0., 5., 0.2)

# plt.plot(x, y, 모양, x, y, 모양, x, y, 모양)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

### 특정 키워드를 사용해서 plt 사용해보기

```python
# np.arange(start, stop) np.arange(50)은 [0, 1, ..., 49]을 의미
# np.random.randint(low, high, size) 아래는 0~49사이의 숫자를 50개 뽑는다는 의미
# np.random.randn(d0,d1 ...) np.random.randn(50)은 0과 1사이의 숫자를 50개 뽑는다는 의미임
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

# plt.scatter(x, y, c=color, s=shape, data=data)
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

### Categorical Variables 를 활용해 plt 사용하기

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9,3))

'''
subplots()와 subplot(int, int, index)의 차이점
subplots는 그래프를 랜덤하게 설정해주는데 subplot(index)는 특정 부분에 subplot을 만들라고 선언함
'''
plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

### 다양한 figures와 axes를 다뤄보자

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])


plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot(111) by default

plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(211)             # make subplot(211) in figure1 current
plt.title('Easy as 1, 2, 3') # subplot 211 title
```

### Text를 활용해 plt 사용하기

```python
mu, sigma = 100, 15
x = mu +sigma*np.random.randn(10000)

n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)

# plt.axis([xmin, xmax, ymin, ymax])
# plt.text(x위치, y위치, 적을 내용, **kwargs 특성들)
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40,160,0,0.03])
plt.grid(True)
plt.show()
```

#### 텍스트 주석달기

```python
ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

# plt.annotate(text, xy=(시작점 위치), xytext=(텍스트 위치), **kwargs arrow의 특성들)
plt.annotate('local max', xy=(2,1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05),)
plt.ylim(-2, 2)
plt.show()
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

# make up some data in the open interval (0, 1)
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure()

'''
plt.yscale(value, **kwargs)
value={'linear', 'log', 'symlog', 'logit', ...}
y축의 범위 및 모양을 나타냄
'''

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthresh=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()
```
