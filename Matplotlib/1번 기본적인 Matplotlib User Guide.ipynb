```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# fig는 단순한 그래프를 의미함 ax라는 것은 하나의 x축과 y축을 생성하는 것임
# ax.plot(x요소, y요소)
# subplots(nrows=number of row, ncols=number of column)으로 이루어짐
fig, ax = plt.subplots()
ax.plot([1,2,3,4], [1,4,2,3])
```

```python
# 이렇게 선언되는 경우 plot(지도를 그리다는 의미)함수는 불러오지 않는다. 
# plot([1,2,3,4],[1,4,2,3])
# 따라서 plt를 선언해줘야 아래와 같이 나올 수 있음
plt.plot([1,2,3,4],[1,4,2,3])
```

#### 용어 설명
1. Figure: 말 그대로 그림이 그려지는 지도 같은 의미
2. Axes: Axis라는 하나의 축이 2개 또는 3개로 이루어짐
3. Legend: 범례라는 의미로 선의 이름을 의미함
4. Major tick label: 위 1.0, 1.5, 2.0, 2.5, 3.0과 같이 큰 요소를 의미
5. Minor tick label: 여기서 1.1, 1.2 이런 것들을 의미
6. X, Y axis label: 각 축의 이름

```python
# np.linspace(start, end, num=50)으로 이루어져 start에서 end까지 50개로 나눈 배열
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots()
ax.plot(x, x, label="linear")
ax.plot(x, x**2, label="quadratic")
ax.plot(x, x**3, label="cubic")
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Simple Plot')
ax.legend()
```

#### 위 요소와 아래요소의 차이점
- 위 요소는 fig, ax = plt.subplots()를 선언했고
- 아래 요소는 그거를 선언하지 않고 plt를 사용해 그린 것임

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

```python
# 이런 식으로 Django에 띄우면 될 것 같음
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph

    Parameters
    ----------
    ax : Axes
        The axes to draw to

    data1 : array
       The x data

    data2 : array
       The y data

    param_dict : dict
       Dictionary of kwargs to pass to ax.plot

    Returns
    -------
    out : list
        list of artists added
    """
    out = ax.plot(data1, data2, **param_dict)
    return out
```

```python
data1, data2, data3, data4 = np.random.randn(4, 100)
fig, ax = plt.subplots(1, 1)
my_plotter(ax, data1, data2, {'marker': 'x'})
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})
```
